## Transactions

```python
#Import des fichiers de transactions
t1= pd.read_csv('MavenMarket_Transactions_1997.csv')
t2= pd.read_csv('MavenMarket_Transactions_1998.csv')
```

```python
t1.head()
```

```python
t1.info()
```

```python
t2.info()
```

```python
#Conversion des dates
t1['transaction_date']= pd.to_datetime(t1['transaction_date'])
t1['stock_date']= pd.to_datetime(t1['stock_date'])
t2['transaction_date']= pd.to_datetime(t2['transaction_date'])
t2['stock_date']= pd.to_datetime(t2['stock_date'])
```

```python
#Renommage des colonnes
t1= t1.rename(columns={'transaction_date':'date_transaction', 'stock_date':'date_stock', 'product_id':'id_produit', 'customer_id':'id_client', 'store_id':'id_magasin', 'quantity':'quantite'})
t2= t2.rename(columns={'transaction_date':'date_transaction', 'stock_date':'date_stock', 'product_id':'id_produit', 'customer_id':'id_client', 'store_id':'id_magasin', 'quantity':'quantite'})
```

```python
t2.head()
```

```python
#Regroupement des 2 fichiers
transac= t1.append(t2, ignore_index=True)
transac
```
